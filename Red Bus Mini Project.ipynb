```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome()

bus_link = [
    'https://www.redbus.in/search?fromCityName=Kozhikode&fromCityId=74661&toCityName=Thiruvananthapuram&toCityId=71425&busType=Any&opId=28011&rDoj=&returnDoj=&isReturn=false&onward=18-Oct-2024',
    'https://www.redbus.in/bus-tickets/kakinada-to-visakhapatnam?fromCityName=Kakinada&fromCityId=316&toCityName=Visakhapatnam&toCityId=248&busType=Any&opId=10283&onward=18-Oct-2024',
    'https://www.redbus.in/bus-tickets/sikar-to-jaipur?fromCityName=Sikar&fromCityId=79729&toCityName=Jaipur%20(Rajasthan)&toCityId=807&busType=Any&opId=15499&onward=18-Oct-2024',
    'https://www.redbus.in/bus-tickets/hyderabad-to-vijayawada?fromCityName=Hyderabad&fromCityId=124&toCityName=Vijayawada&toCityId=134&busType=Any&opId=18491&onward=18-Oct-2024',
    'https://www.redbus.in/bus-tickets/kolkata-to-durgapur?fromCityName=Kolkata&fromCityId=74820&toCityName=Durgapur%20(West%20Bengal)&toCityId=69802&busType=Any&opId=16374&onward=18-Oct-2024',
    'https://www.redbus.in/bus-tickets/delhi-to-chandigarh?fromCityName=Delhi&fromCityId=733&toCityName=Chandigarh&toCityId=833&busType=Any&opId=16227&onward=18-Oct-2024',
    'https://www.redbus.in/bus-tickets/delhi-to-bareilly?fromCityName=Delhi&fromCityId=733&toCityName=Bareilly%20(Uttar%20Pradesh)&toCityId=1437&busType=Any&opId=25946&onward=18-Oct-2024',
    'https://www.redbus.in/bus-tickets/digha-to-barasat-west-bengal?fromCityName=Digha&fromCityId=74706&toCityName=Barasat%20(West%20Bengal)&toCityId=201668&busType=Any&opId=16426&onward=18-Oct-2024',
    'https://www.redbus.in/bus-tickets/ludhiana-to-chandigarh?fromCityName=Ludhiana&fromCityId=736&toCityName=Chandigarh&toCityId=833&busType=Any&opId=27455&onward=18-Oct-2024',
    'https://www.redbus.in/bus-tickets/delhi-to-patiala?fromCityName=Delhi&fromCityId=733&toCityName=Patiala&toCityId=82742&busType=Any&opId=11060&onward=18-Oct-2024'
]

bus_data = []
limit = 100

current_link_index = 0

while current_link_index < len(bus_link):
    driver.get(bus_link[current_link_index])
    time.sleep(5)
    
    start_time = time.time()

    current_link_data = []

    while len(current_link_data) < limit:
        bus_route = driver.find_elements(By.XPATH, "//div[@class='h2-tag-seo']")
        bus_names = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
        bus_types = driver.find_elements(By.XPATH, "//div[contains(@class, 'bus-type')]")
        depart_times = driver.find_elements(By.XPATH, "//div[@class='dp-time f-19 d-color f-bold']")
        durations = driver.find_elements(By.XPATH, "//div[@class='dur l-color lh-24']")
        reaching_times = driver.find_elements(By.XPATH, "//div[@class='bp-time f-19 d-color disp-Inline']")
        star_ratings = driver.find_elements(By.XPATH, "//div[@class='lh-18 rating rat-green ']")
        prices = driver.find_elements(By.XPATH, "//div[@class='fare d-block']")
        seat_availabilities = driver.find_elements(By.XPATH, "//div[@class='seat-left m-top-16']")
    
        for i in range(len(bus_names)):
            current_link_data.append({
                'link': bus_link[current_link_index],
                'route_name': bus_route[i].text if i < len(bus_route) else None,
                'name': bus_names[i].text if i < len(bus_names) else None,
                'bus_type': bus_types[i].text if i < len(bus_types) else None,
                'departure_time': depart_times[i].text if i < len(depart_times) else None,
                'duration': durations[i].text if i < len(durations) else None,
                'reaching_time': reaching_times[i].text if i < len(reaching_times) else None,
                'star_rating': star_ratings[i].text if i < len(star_ratings) else None,
                'price': prices[i].text if i < len(prices) else None,
                'seat_availability': seat_availabilities[i].text if i < len(seat_availabilities) else None,
            })
    
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
        time.sleep(3)
    
        if time.time() - start_time >= 5:
            print(f"Number of bus data fetched for link {current_link_index + 1}: {len(current_link_data)}")
            start_time = time.time()

    bus_data.extend(current_link_data)

    current_link_index += 1

for bus in bus_data:
    print(bus)

driver.quit()
```

```python
import mysql.connector

# Connect to the MySQL database
db_connection = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="23569",
    database="red_bus"
)

cursor = db_connection.cursor()

# Create the bus_info table
create_table_query = """
CREATE TABLE IF NOT EXISTS bus_info (
    id INT AUTO_INCREMENT PRIMARY KEY,
    route_name VARCHAR(255),
    name VARCHAR(255),
    bus_type VARCHAR(100),
    departure_time VARCHAR(50),
    duration VARCHAR(50),
    reaching_time VARCHAR(50),
    star_rating VARCHAR(10),
    price VARCHAR(50),
    seat_availability VARCHAR(50),
    link VARCHAR(255) NOT NULL
);
"""

cursor.execute(create_table_query)
db_connection.commit()
```
